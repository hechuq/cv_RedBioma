**A. Generacion de todas las composiciones**

```python
import numpy as np
import cv2 as cv
import os
import glob

# Google Colab requiere usar esta funcion.
# Se indica que la funcion cv.imshow no esta admitida.
# Se incluye cv.imshow en linea comentada
from google.colab.patches import cv2_imshow
```

```python
# Directorio para composiciones
directory = "merged_images"

# Revisar existencia de carpeta y crearla.
if not os.path.exists(directory):
    os.makedirs(directory, exist_ok=True)
    print(f"Carpeta '{directory}' creado.")
else:
    print(f"Carpeta '{directory}' ya existe.")
```

```python
files = sorted(glob.glob("originales/*"))

for img_file in files:
  # Leyendo imagen y extrayendo dimensiones
  img = cv.imread(img_file)
  print(f"Imagen '{img_file}' con dimensiones: alto = {img.shape[0]}, ancho = {img.shape[1]} ")

  # Separacion de canales y generacion de imagenes en azul, verde y rojo
  b, g, r = cv.split(img)

  zeros = np.zeros(img.shape[:2], dtype=np.uint8)
  b_img = cv.merge([b, zeros, zeros])
  g_img = cv.merge([zeros, g, zeros])
  r_img = cv.merge([zeros, zeros, r])

  # Composicion de imagenes azul, verde, rojo y original
  composite = cv.hconcat([b_img, g_img, r_img, img])
  print(f"  Generacion de imagen compuesta OK.")

  # Creacion de ruta para imagen compuesta
  img_base = os.path.basename(img_file)
  img_name, img_ext = os.path.splitext(img_base)
  img_type = "_merged_"
  mer_date = "20250630"
  merged_path = os.path.join(directory, img_name + img_type + date_yymmdd + img_ext)
  print(f"  Ruta para composicion: '{merged_path}'")

  # Guardar imagen compuesta en nueva ruta
  cv.imwrite(merged_path, composite)
  img_merged = cv.imread(merged_path)
  img_size = img_merged.shape[:2]
  print(f"  Composicion guardada. Con dimensiones: alto = {img_size[0]}, ancho = {img_size[1]} ")
```

```python
# Verificando imagenes compuestas
merged_files = sorted(glob.glob("merged_images/*"))

for m in merged_files:
  img_comp = cv.imread(m)

  # Factor de reduccion con altura fija de 200
  h, w = img_comp.shape[:2]
  new_h = 200
  fac_size = new_h / h
  new_w = int(w * fac_size)
  new_dim = (new_w, new_h)

  # Reduccion de imagenes compuestas y visualizacion
  img_small = cv.resize(img_comp, new_dim)
  print(os.path.basename(m))
  cv2_imshow(img_small)
  #cv.imshow(img_small)
```
